```python
q1=np.percentile(df['Age'],25)
q1
```

```python
iqr=q3-q1
iqr
```

```python
LBR=q1-(1.5*iqr)
LBR
```

```python
UBR=q3+(1.5*iqr)
UBR
```

```python
df1=df[(df['Age']<LBR) | (df['Age']>UBR)]
df1
```
