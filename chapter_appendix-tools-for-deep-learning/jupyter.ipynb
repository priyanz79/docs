```python
import statistics
import math
n=int(input("Enter the amount of numbers: "))
z = []
for _ in range(n):
  num = int(input("Enter a number: "))
  z.append(num)
print(z)
maximum=max(z)
print("Max No.",maximum)
minimum=min(z)
print("Min No.",minimum)
total_sum = sum(z)
print("Sum of the elements: ", total_sum)
average = total_sum / n
print("Average of the elements: ", average)
sqaure_root = math.sqrt(total_sum)
print("Square root of the elements: ", sqaure_root)]
```
